```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

num_time_steps = 100
num_shelters = 5

time_steps = np.arange(num_time_steps)
evacuated_residents = np.cumsum(np.random.randint(0, 10, num_time_steps))
shelter_usage = np.random.randint(0, 100, (num_time_steps, num_shelters))

shelter_usage_df = pd.DataFrame(shelter_usage, columns=[f"Shelter {i+1}" for i in range(num_shelters)])

cyclone_path_x = np.cumsum(np.random.choice([-1, 0, 1], num_time_steps))
cyclone_path_y = np.cumsum(np.random.choice([-1, 0, 1], num_time_steps))

evacuated_residents, shelter_usage_df.head(), cyclone_path_x, cyclone_path_y
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time_steps, evacuated_residents, color='blue', marker='o')
plt.title("Number of Evacuated Residents Over Time")
plt.xlabel("Time Steps")
plt.ylabel("Evacuated Residents")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
for column in shelter_usage_df.columns:
    plt.plot(time_steps, shelter_usage_df[column], label=column)
plt.title("Shelter Usage Over Time")
plt.xlabel("Time Steps")
plt.ylabel("Shelter Occupancy")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(cyclone_path_x, cyclone_path_y, color='red', marker='x')
plt.title("Cyclone Path")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.grid(True)
plt.show()
```
